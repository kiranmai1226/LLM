```python
import json
import os

def clean_notebook(notebook_path):
    """
    Clean notebook metadata (especially widget data) to prevent errors.
    """
    if not os.path.exists(notebook_path):
        print(f"Notebook not found: {notebook_path}")
        return

    # Open and load the notebook
    with open(notebook_path, 'r') as f:
        notebook_content = json.load(f)

    # Remove 'widgets' metadata if it exists
    if 'widgets' in notebook_content['metadata']:
        print("Removing 'widgets' metadata...")
        del notebook_content['metadata']['widgets']

    # Save the cleaned notebook to the same location
    with open(notebook_path, 'w') as f:
        json.dump(notebook_content, f)

    print(f"Cleaned notebook saved as '{notebook_path}'.")


# Example of running it on the notebook from Google Drive
notebook_path = '/content/drive/MyDrive/Colab Notebooks/huggingface_transformers.ipynb'
clean_notebook(notebook_path)
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

```python
tokenizer=AutoTokenizer.from_pretrained("google/flan-t5-large")
```

```python
model=AutoModelForSequenceClassification.from_pretrained("google/flan-t5-large")
```

```python
classifier = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer,device=0)

res = classifier("I've been waiting for a HuggingFace course my whole life.")

print(res)
```
