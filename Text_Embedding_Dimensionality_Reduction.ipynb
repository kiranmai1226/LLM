```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.stem import WordNetLemmatizer
story=[]
lemmatizer=WordNetLemmatizer()
for sent in raw_sent:
  lemmatized_sentance=[lemmatizer.lemmatize(word) for word in simple_preprocess(sent)]
  story.append(lemmatized_sentance)
```

```python
from nltk.corpus import stopwords
import nltk
nltk.download("stopwords")
```

```python
stop_words = set(stopwords.words('english'))
```

```python
filtered_story=[]
def remove_stopwords(story):

    for sentence in story:
        filtered_sentence = [word for word in sentence if word.lower() not in stop_words]
        filtered_story.append(filtered_sentence)
    return filtered_story
```

```python
remove_stopwords(story)
```

```python
filtered_story[0]
```

```python
custom_model=gensim.models.Word2Vec(window=25,min_count=2,vector_size=700,epochs=10)
```

```python
custom_model.build_vocab(filtered_story)
```

```python
custom_model.train(filtered_story,total_examples=custom_model.corpus_count,epochs=custom_model.epochs)
```

```python
vocab=list(custom_model.wv.index_to_key)
embeddings=custom_model.wv[vocab]
```

```python
import os
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
pca=PCA(n_components=2)
result=pca.fit_transform(embeddings)
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)
print(cumulative_variance)
similarity_matrix = cosine_similarity(result)
print(similarity_matrix[0,10])
plt.figure(figsize=(10,10))
plt.scatter(result[:,0],result[:,1])
#for i, word in enumerate(vocab):
  #plt.annotate(word,xy=(result[i,0],result[i,1]))
plt.title("PCA visualization of word embeddings")
plt.show()


```
